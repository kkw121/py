### 2. These graphs gives us an idea for how the petal_length and petal_width varies between the three iris species


```python
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(font_scale=0.7, style="white", rc={"axes.facecolor": (0, 0, 0, 0)})


pal = sns.cubehelix_palette(3, rot=-.25, light=.7)
g = sns.FacetGrid(irid, row="species", hue="species", aspect=15, height=1, palette=pal)

g.map(sns.kdeplot, "petal_length",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "petal_length", clip_on=False, color="w", lw=1, bw_adjust=.5)

g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)

def label(x, color, label):
    ax = plt.gca()
    ax.text(0, 0.1, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "petal_length")

g.figure.subplots_adjust(hspace=-.75)

g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(font_scale=0.7, style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

pal = sns.cubehelix_palette(3, rot=-.25, light=.7)
g = sns.FacetGrid(irid, row="species", hue="species", aspect=15, height=1, palette=pal)

g.map(sns.kdeplot, "petal_width",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "petal_width", clip_on=False, color="w", lw=1, bw_adjust=.5)

g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)

def label(x, color, label):
    ax = plt.gca()
    ax.text(0, 0.1, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "petal_width")

g.figure.subplots_adjust(hspace=-.75)

g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
```

### 3. The pairplot visualization of the Iris dataset reveals distinct clusters among the species, particularly evident in the scatterplots of petal length versus petal width, where setosa appears clearly separated from the other species. Additionally, there is noticeable overlap between versicolor and virginica, especially in sepal length versus sepal width. The histograms along the diagonal show variations in feature distributions, notably indicating that setosa tends to have smaller dimensions compared to the other species. These observations suggest that petal dimensions are more effective in distinguishing between iris species compared to sepal dimensions

```python
from sklearn.datasets import load_iris

iris = load_iris()
irid = pd.DataFrame(data=iris.data, columns=iris.feature_names)
irid['species'] = iris.target

# Map target integers to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
irid['species'] = irid['species'].map(species_map)

# Create pairplot
sns.pairplot(irid, hue="species")
plt.show()
```

## Data Visualization 2

> Visualization for Geyser Dataset



```python
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

### 1. The pairplot shows scatterplots of eruption duration versus waiting time, revealing two distinct clusters or groups of data points, suggesting potential classes or patterns in the dataset. The histograms along the diagonal show bimodal distributions for both eruption duration and waiting time, indicating the presence of two distinct groups within the dataset.


```python
# Load the Geyser dataset
geyser_df = sns.load_dataset("geyser")

# Create pairplot
sns.pairplot(geyser_df)
plt.show()
```

### 2. The histogram of eruption duration displays two peaks, suggesting the presence of two distinct types of eruptions with different durations. The histogram of waiting time also exhibits two peaks, indicating two distinct waiting time periods between eruptions.

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(geyser_df["duration"], kde=True)
plt.title("Eruption Duration")
plt.subplot(1, 2, 2)
sns.histplot(geyser_df["waiting"], kde=True)
plt.title("Waiting Time")
plt.tight_layout()
plt.show()
```

### 3. The hexbin plot provides a clearer visualization of the density of data points in the scatterplot, indicating areas of high and low data density. It confirms the clustering observed in the pairplot and highlights regions with a higher concentration of data points.


```python
sns.jointplot(x="duration", y="waiting", data=geyser_df, kind="hex")
plt.show()
```
